```python
import numpy as np



mod = 10

#fib short for fibonacci sequence
def fib(n):
    fib = [1,5]
    for i in range(2,n):
        next_term = (fib [i-1] + fib [i-2]) % mod
        fib.append(next_term)
    return fib

print(fib(60))
print(f'length of sequence is',len(fib(60)))
```

```python
#initial brute force method using sum of two numbers make 100

mod = 10
def fibonacci(sum_range, n):
    results = set()  
    for total_sum in range(sum_range[0], sum_range[1] + 1):
        for i in range(10):  
            j = total_sum - i
            if 0 <= j <= 9:  
                fib = [i , j]
                for k in range(2, n):
                    next_term = (fib[k - 1] + fib[k - 2]) % mod
                    fib.append(next_term)

                results.add(tuple(fib))  

    print(f"Total unique sequences: {len(results)}")
    return [list(sequence) for sequence in results]

# To consider other mod values change the j range and sum ranges and number of terms 
sum_range = [0, 18]
num_terms = 60
result = fibonacci(sum_range, num_terms)

for sequence in result:
    print(f"{sequence[0]} and {sequence[1]}: {sequence[2:]}")
    
```
